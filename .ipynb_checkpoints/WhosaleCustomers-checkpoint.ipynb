```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Punctul 1: Încarcărea și analiza setului de date
data = pd.read_csv('data/data.csv')

# Afișează primele câteva rânduri ale setului de date
print(data.head())

# Verifică dacă există valori lipsă
print(data.isnull().sum())
```

```python
# Punctul 2: Preprocesarea datelor
# Elimină coloanele 'Channel' și 'Region' deoarece sunt categorice și nu sunt necesare pentru clustering
data = data.drop(['Channel', 'Region'], axis=1)

# Normalizează datele
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Punctul 3.1: Aplicarea algoritmului K-Means
# Determină numărul optim de clustere folosind metoda Elbow
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

# Trasează graficul metodei Elbow
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Metoda Elbow')
plt.xlabel('Numărul de clustere')
plt.ylabel('WCSS')
plt.show()
```

```python
# Bazat pe metoda Elbow, alege numărul optim de clustere (6)
kmeans = KMeans(n_clusters=6, init='k-means++', random_state=42)
kmeans.fit(data_scaled)
data['KMeans_Cluster'] = kmeans.labels_
```

```python
# Punctul 3.2: Aplicarea algoritmului DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(data_scaled)
data['DBSCAN_Cluster'] = dbscan.labels_
```

```python
# Punctul 4: Evaluarea clusteringului K-Means folosind Scorul Silhouette
silhouette_avg = silhouette_score(data_scaled, kmeans.labels_)
print(f'Scorul Silhouette pentru K-Means: {silhouette_avg}')

# Evaluarea clusteringului DBSCAN folosind Scorul Silhouette
silhouette_avg_dbscan = silhouette_score(data_scaled, dbscan.labels_)
print(f'Scorul Silhouette pentru DBSCAN: {silhouette_avg_dbscan}')
```

```python
# Punctul 5.1: Interpretarea clusterelor
# Analizează valorile medii ale fiecărei caracteristici pentru clusterele K-Means
cluster_analysis = data.groupby('KMeans_Cluster').mean()
print(cluster_analysis)

# Vizualizarea clusterelor (pentru vizualizare 2D, folosește PCA pentru reducerea dimensionalității)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)
data['PCA1'] = data_pca[:, 0]
data['PCA2'] = data_pca[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='KMeans_Cluster', data=data, palette='viridis')
plt.title('Clustering K-Means cu PCA')
plt.show()
```

```python
# Punctul 5.2: Recomandări bazate pe analiza clusterelor
# Exemplu: Identifică clusterele cu cheltuieli mari și sugerează strategii de marketing direcționate
high_spending_cluster = cluster_analysis.loc[cluster_analysis['Fresh'].idxmax()]
print(f"Caracteristicile clusterului cu cheltuieli mari:\n{high_spending_cluster}")
```
